```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head(20)
```

```python
x = insurance_data.iloc[:,0].values
y = insurance_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
age_model = LinearRegression()
age_model.fit(x,y)
```

```python
y_pred = age_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
unk_x = [[80]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = age_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()

print("Insurance charges should be:", age_model.predict(unk_x))
```
